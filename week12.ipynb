```python
#12주차 실습1: 서울의 일교차

import csv
f=open('seoul.csv','r',encoding='cp949')
data=csv.reader(f)
next(data) #헤더 건너뜀
max_diff=0
diffmax_date=''
min_temp=999
min_date=''
max_temp=-999
max_date=''
for row in data:
    if row[-1]=='' or row[-2]=='':
        row[-1]=-999
        row[-2]=999
    row[-1]=float(row[-1])
    row[-2]=float(row[-2])
    tempdiff=row[-1]-row[-2] #최고기온-최저기온
    if tempdiff > max_diff:
        max_diff=tempdiff
        diffmax_date=row[0] #일교차가 큰 날짜 저장
    if min_temp>row[-2]: #최저기온 
        min_date=row[0]
        min_temp=row[-2]  
    if max_temp<row[-1]: #최고기온 
        max_date=row[0]
        max_temp=row[-1]

f.close()
print("일교차가 가장 큰 날\n",diffmax_date) 
print("최저기온이 가장 낮은 날\n",min_date)
print("최고기온이 가장 높은 날\n",max_date)
```
